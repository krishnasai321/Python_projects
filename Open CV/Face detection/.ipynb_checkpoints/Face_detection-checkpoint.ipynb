```python
#Importing libraries
import cv2
import os
import matplotlib.pyplot as plt
```

```python
#Setting directory
os.chdir('C:\\Data\\github\\python_projects\\Open CV\\Face detection')
```

```python
#Using the frontalface xml file
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
face_cascade.empty()
```

```python
#Selecting original image
img = cv2.imread('sachin-tendulkar.jpg')
```

```python
# Required to plot images in notebooks
%matplotlib inline
```

```python
#By default, matplot lib plot will not be in RGB. Covert to RGB before plotting
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
#Detection corners of face
faces = face_cascade.detectMultiScale(img, 1.1, 4)

#Creating a rectangle
for (x, y, w, h) in faces: 
  cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)
```

```python
#Plotting the new photo with detected face highlighted
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
#Saving the image
cv2.imwrite("sachin_face_detected.png", img)
```
