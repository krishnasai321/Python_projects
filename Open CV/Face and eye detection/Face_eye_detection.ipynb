```python
#Importing libraries
import cv2
import os
import matplotlib.pyplot as plt
```

```python
#Setting directory
os.chdir('C:\\Data\\github\\python_projects\\Open CV\\Face and eye detection')
```

```python
#Using the frontalface and eye xml file
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')

print(face_cascade.empty())
print(eye_cascade.empty())
```

```python
#Selecting original image
img = cv2.imread('sachin-tendulkar.jpg')
```

```python
# Required to plot images in notebooks
%matplotlib inline
```

```python
#By default, matplot lib plot will not be in RGB. Covert to RGB before plotting
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

#Detection corners of face and eyes
faces = face_cascade.detectMultiScale(gray, 1.3, 5)
for (x,y,w,h) in faces:
     cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
     roi_gray = gray[y:y+h, x:x+w]
     roi_color = img[y:y+h, x:x+w]
     eyes = eye_cascade.detectMultiScale(roi_gray)
     for (ex,ey,ew,eh) in eyes:
         cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)
```

```python
#Plotting the new photo with detected face highlighted
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
#Saving the image
cv2.imwrite("sachin_face_eye_detected.png", img)
```
